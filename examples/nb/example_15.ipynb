# Oceanic Ni√±o Index (ONI) lineplot

We have our own little function to plot ONI lines. This function uses the output data frame from our `get_oni()` function in the `obrero.experimental.enso` module. First let's import obrero:

```python
# small hack to be able to import module without install
import os
import sys
sys.path.append(os.getcwd() + '/../')

import obrero
import obrero.plot as oplot
from obrero.experimental import enso
```

We will plot ONI calculated from the NOAA anomalies data and from an experiment. Read data file with sea surface temperature (SST) values:

```python
# filenames
f1 = 'data/sst_ctl.nc'
f2 = 'data/sst_amp50.nc'
f3 = 'data/sst_amp300.nc'

# read as datarray
damip = obrero.read_nc(f1, 'tosbcs')
damp50 = obrero.read_nc(f2, 'tosbcs')
damp300 = obrero.read_nc(f3, 'tosbcs')
```

Compute all ONIs:

```python
noaa = enso.get_oni(years=range(2005,2011))
amip = enso.get_oni(damip)
amp50 = enso.get_oni(damp50)
amp300 = enso.get_oni(damp300)
```

And now comes the plotting. We create a list with data and then a dictionary with plotting specifications:

```python
%matplotlib inline
dlist = [noaa, amip, amp50, amp300]
spec = dict(names=['ERSSTv5', '1.0 AMIP-II', '1.5 AMIP-II', '3.0 AMIP-II'],
            colors=['black', 'OrangeRed', 'FireBrick', 'Maroon'],
            ylim=[-8, 8])
ax = oplot.oni_lineplot(dlist, **spec)
```

We can see how input SST for different experiments varies.
