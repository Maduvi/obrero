# Plotting binary masks: land-sea and glacier

This set of tools included in obrero were made thinking in a specific climate model. So it is no wonder we have a sort of "specific" kind of tools. These are for instance `plot_landsea()` and `plot_glacier()` functions. Though they can also be used for other arrays, they were created with a specific model in mind. These could be replaced with our very own more generic `plot_global_contour()` calls but we simply wanted these two additional perks of not needing a lot code to plot these. Import obrero:

```python
# small hack to be able to import module without install
import os
import sys
sys.path.append(os.getcwd() + '/../')

import obrero
```

Now read land binary mask and glacier cover:

```python
# file names
f1 = 'data/lsm.nc'
f2 = 'data/glac.nc'

# read data
lsm = obrero.read_nc(f1, 'lsm')
glac = obrero.read_nc(f2, 'glac')

# choose single time (they never change)
lm = lsm[0]
gl = glac[0]
```

We still need to mask these two arrays to be land-only:

```python
# mask out ocean
land_lm = obrero.mask_land_ocean(lm, lm)
land_gl = obrero.mask_land_ocean(gl, lm)
```

And now we plot using the functions:

```python
%matplotlib inline
ax = obrero.plot_landsea(land_lm)
```

```python
ax = obrero.plot_glacier(land_gl)
```

Notice these functions return axes, in case the user wants to put these maps elsewhere. Additional options for these functions are similar to those for `panel_global_contour()`.
